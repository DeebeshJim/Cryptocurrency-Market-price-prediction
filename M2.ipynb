# DATA VISUALIZATION AND DATA ANALYSIS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read the datasets
data=pd.read_csv('Bitcoin.csv')
del data['SNo']
del data['Name']
del data['Symbol']
del data['Date']
del data['Marketcap']
data.head()
```

```python
data.columns
```

```python
plt.figure(figsize=(12,7))
sns.countplot(x='Close',data=data)
```

```python
plt.figure(figsize=(20,10))

plt.subplot(1,2,1)
plt.hist(data['Open'],color='blue')

plt.subplot(1,2,2)
plt.hist(data['Close'],color='green')
```

```python
data.hist(figsize=(15,60),layout=(15,4),color='green')
plt.show()
```

```python
plt.boxplot(data['Open'])
```

```python
sns.pairplot(data)
```

```python
fig,ax=plt.subplots(figsize=(20,15))
sns.heatmap(data.corr(),annot=True,fmt='0.2%',cmap='autumn',ax=ax)
```
