## DATA VISUALIZATION AND DATA ANALYSIS

```python
# Import necessary libraries.
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Avoid unnessary warnings in this module(Ex: software updation, version-mismatch, cache errors)
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read the datasets.
# Read the datasets
df=pd.read_csv('Bitcoin.csv')
del df['SNo']
del df['Name']
del df['Symbol']
del df['Date']
del df['Marketcap']
```

```python
# Check the head values.
df.head()
```

```python
# Check our datasets information (ex: data_type, NUlL_count, Columns, and so on)
df.info()
```

```python
# Remove the null data
df=df.dropna()
```

```python
# Check the bottom value of this datasets 
df.tail()
```

```python
# Plot a Histogram.
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.hist(df['High'],color='DarkOrange')

plt.subplot(1,2,2)
plt.hist(df['Low'],color='HotPink')
```

```python
# Plot the Histogram
df.hist(figsize=(15,55),layout=(15,4), color='MistyRose')
plt.show()
```

```python
# Plot the Histogram
df['Open'].hist(figsize=(10,5),color='Magenta')
```

```python
# Scatter Plot
plt.scatter(df['Close'],df['Open'], color='darkslategray') # scatter, plot, triplot, stackplot
```

```python
# Check the outlayers of this datasets
plt.boxplot(df[''], boxprops={'color': 'Violet'})
plt.show()
```

```python
# Plot a density graph
df['T amb (Degree)'].plot(kind='density',color="Plum")
```

```python
# Violin plot
sns.violinplot(df['Wind Speed at 76.8 m\n(m/sec)'], color='Plum') 
# barplot, boxenplot, boxplot, countplot, displot, distplot, ecdfplot, histplot, kdeplot, pointplot, violinplot, stripplot
```

```python
# Check the relation between each columns using heatmap.
fig, ax = plt.subplots(figsize=(20,15))
sns.heatmap(df.corr(),annot = True, fmt='0.2%',cmap = 'autumn',ax=ax)
```

```python
# Visualise the distribution of data.
sns.histplot(df['Power Generated\n(kw)'], kde=True)
plt.xlabel('Power Generated\n(kw)')
plt.ylabel('Frequency')
plt.title('Distribution Plot of Power Generated')
plt.show()
```
