# XGBOOST REGRESSOR ALGORITHEM

```python
# Import the necessary libraries
import pandas as pd
import numpy as np
```

```python
# Avoid unnessary warnings in this module(Ex: software updation, version-mismatch, cache errors) 
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read the datasets
df=pd.read_csv('USDCoin.csv')
```

```python
# Check the head values.
df.head()
```

```python
# Check our data how many columns they are.
df.columns
```

```python
del df['SNo']
del df['Name']
del df['Symbol']
del df['Date']
del df['Marketcap']
```

```python
# Drop the null values of in this datasets.
df1=df.dropna()
```

```python
# Check the bottom line records of this data.
df1.tail()
```

```python
# Split the datasets features into two varibles (dependent, independent)
# Independent variables
x = df.drop(labels='Power_Generated_kw', axis=1)

# Dependent variable
y = df.loc[:,'Power_Generated_kw']
```

```python
x
```

```python
# Split the data sets into two methods Training, Testing 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
print("NUMBER OF TRAIN DATASET    : ", len(x_train))
print("NUMBER OF TEST DATASET      : ", len(x_test))
print("TOTAL NUMBER OF DATASET    : ", len(x_train)+len(x_test))
```

```python
# # Build a Learning pattern algorithm for regression problem.
# import xgboost as xgb
from sklearn.ensemble import RandomForestRegressor
```

```python
XGB = RandomForestRegressor()
# Fit is the training funtion execute to learn
XGB.fit(x_train, y_train)
```

```python
# Test the input features.
y_pred = XGB.predict(x_test)
```

```python
# Check the Rand score
from sklearn.metrics import rand_score
RAND = rand_score(y_test, y_pred)
print("MEAN RAND SCOORE OF XG BOOST REGRESSOR",RAND*100)
```

```python
# Checkout the mean squared error rate
from sklearn.metrics import mean_squared_error
MSE = mean_squared_error(y_test, y_pred)
print("MEAN SQUARE ERROR SCOORE OF XG BOOST REGRESSOR",MSE*100)
```

```python
# Checkout the mean absolute error
from sklearn.metrics import mean_absolute_error
MAE = mean_absolute_error(y_test, y_pred)
print("MEAN ABSOLUTE ERROR OF XG BOOST REGRESSOR",MAE)
```

```python
# Checkout the R2score
from sklearn.metrics import r2_score
R2 = r2_score(y_test, y_pred)
print("R-SQUARE VALUE OF XG BOOST REGRESSOR",R2)
```

```python
# Save the model for deployment purpose.
import joblib
joblib.dump(XGB, 'Generation2.pkl')
```
